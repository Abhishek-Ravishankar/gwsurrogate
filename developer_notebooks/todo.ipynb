### Physical waveform evaluations (any permissible input)

```python
### python interface to EOBNRv2-LAL ###
def EOBNRv2_LAL(Mtot,q,dt,fmin,Dist,inclination):
        
    M1,M2  = gwtools.Mq_to_m1m2(Mtot,q)
    hp, hc = lalsim.SimInspiralChooseTDWaveform(phiRef=0.0,deltaT=dt,m1=M1,m2=M2,s1x=0.0,s1y=0.0,s1z=0.0,\
                                                s2x=0.0,s2y=0.0,s2z=0.0,f_min=fmin,f_ref=0.0,\
                                                r=Dist,i=inclination,lambda1=0.0,lambda2=0.0,\
                                                waveFlags=None,nonGRparams=None,amplitudeO=0,phaseO=7,\
                                                approximant=lalsim.EOBNRv2)
    h      = hp.data.data + (1j)*hc.data.data
    times  = np.arange(np.size(hp.data.data))*hp.deltaT
    
    times,h = gwtools.remove_amplitude_zero(times,h)
    
    return times, h
```

```python
### comparison for physical waveform modes ###
M    = 80.0
q    = 1.0
incl = 2.0
Dist = 1.0 * 1e6 * lal.PC_SI
Mtot = M * lal.MSUN_SI
fmin = 10.0
dt   = 1.0/2048.0

# LAL evaluation... #
time_LAL, h_LAL = EOBNRv2_LAL(Mtot,q,dt,fmin,Dist,incl)
```

```python
# Surrogate evaluation... #
time_sur,hp,hc = EOBNRv2_sur(q=q,M=M,dist=1.0,f_low=fmin,theta=incl,fake_neg_modes=True)
h_sur          = hp + 1.0j*hc

time_sur,h_sur = gwtools.remove_amplitude_zero(time_sur,h_sur)

### minimize the norm error ###
[tmp, minimized_norm], [tc, phic], [t1,h1,h2] = \
    gwtools.minimize_norm_error(time_sur,h_sur,time_LAL,h_LAL,gwtools.euclidean_norm_sqrd,.0125,.0125,method='nelder-mead')
print "Global minimizer is deltaT = %e and deltaPhi = %e " %(tc,phic)
print "The minimized relative norm error is %e " %minimized_norm

### plot aligned waveforms ###
time_sur_aligned = gwtools.coordinate_time_shift(time_sur,-tc)
h_sur_aligned    = gwtools.modify_phase(h_sur,-phic)

gwtools.plot_pretty(time_LAL,[h_LAL.real,h_LAL.imag],showQ=False)

plt.pyplot.figure(2)
plt.pyplot.plot(time_sur,h_sur,'red')
plt.pyplot.plot(time_LAL,h_LAL,'b--')

plt.pyplot.figure(3)
plt.pyplot.subplot(2,1,1)
plt.pyplot.plot(time_sur_aligned,h_sur_aligned,'red')
plt.pyplot.plot(time_LAL,h_LAL,'b--')
plt.pyplot.xlim([0.5, time_sur[-1]])


### plot pointwise difference ###
t_start, t_end = gwtools.find_common_time_window(time_sur_aligned,time_LAL)
common_times   = np.arange(t_start+.0125,t_end-.0125,(time_LAL[2]-time_LAL[1])) 

from scipy.interpolate import interp1d
h_sur_interp   = interp1d(time_sur_aligned,h_sur_aligned)
h_LAL_interp   = interp1d(time_LAL,h_LAL)

h_sur_eval = h_sur_interp(common_times)
h_LAL_eval = h_LAL_interp(common_times)
err_h = h_sur_eval - h_LAL_eval

plt.pyplot.subplot(2,1,2)
plt.pyplot.semilogy(common_times,np.abs(err_h)/ np.abs(h_LAL_eval),'black')
plt.pyplot.xlim([0.5, time_sur[-1]])
plt.pyplot.title("aligned waveform difference |h1 - h2|/|h1|")

plt.pyplot.show()
```

```python
def waveform_function(x,times=None):

    tc = x[0]
    phic = x[1]

    if times is not None:
        times = times + tc
        times,hp,hc = EOBNRv2_sur(q=q,M=M,dist=1.0,theta=incl,phi=phic,\
                                  samples=times,samples_units='mks',fake_neg_modes=True)
    else:
        times,hp,hc = EOBNRv2_sur(q=q,M=M,dist=1.0,theta=incl,phi=phic,fake_neg_modes=True)
        
    h = hp + 1.0j*hc
    return times, h
```
